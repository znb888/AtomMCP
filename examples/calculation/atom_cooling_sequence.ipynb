````python

import numpy as np
import matplotlib.pyplot as plt

# Define the time steps
steps = 700
time = np.arange(steps)

# Define the parameter arrays
U1 = np.zeros(steps)
U2 = np.zeros(steps)
H = np.zeros(steps)
U = 40

# Stage 1: Lorentz trap (200 steps)
U1[:200] = U
U2[:200] = 0
H[:200] = 0

# Stage 2: Transition to cross trap (300 steps)
U1[200:500] = np.linspace(U, U/2, 300)
U2[200:500] = np.linspace(0, U/2, 300)
H[200:500] = 0

# Stage 3: Evaporative cooling (200 steps)
U1[500:] = U/2
U2[500:] = U/2
H[500:] = np.linspace(0, 1.0, 200) # Assuming H ramps up to 1.0

# Plot the sequences
plt.figure(figsize=(12, 6))
plt.plot(time, U1, label='U1')
plt.plot(time, U2, label='U2')
plt.plot(time, H, label='H (Gradient)')
plt.title('Control Parameter Time Sequence')
plt.xlabel('Time Steps')
plt.ylabel('Parameter Value')
plt.legend()
plt.grid(True)
plt.show()
````
